```python
def finder(x1,f1):
    emp_x1=[]
    emp_f1=[]
    for i_in,i in enumerate(x1):
        for j_in,j in enumerate(i):
            if(j==0):
                emp_x1.append(i_in*10+j_in)
                
    for i_in,i in enumerate(f1):
        for j_in,j in enumerate(i):
            if(j==0):
                emp_f1.append(i_in*10+j_in)
                
    
    return list(set(emp_x1) & set(emp_f1))    
            
```
